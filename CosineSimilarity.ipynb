```python
# Menyimpan masing-masing dataset kedalam variabel

rating = pd.read_csv('tourism_rating.csv')
place = pd.read_csv('tourism_with_id.csv')
user = pd.read_csv('user_new.csv')
```

```python
# Membuang kolom yang tidak dipakai

place = place.drop(['Unnamed: 11','Unnamed: 12'],axis=1)
place = place.drop('Time_Minutes', axis=1)
place.head(2)
```

```python
# Merubah data agar hanya dari Kota Bandung

place = place[place['City']=='Bandung']
place.head(2)
```

```python
# Merubah data rating agar hanya berisi rating pada tempat wisata dari Kota Bandung

rating = pd.merge(rating, place[['Place_Id','Place_Name','Category']], how='right', on='Place_Id')
rating.head()
```

```python
# Merubah data user agar hanya berisi user yang pernah megunjungi wisata di Kota Bandung

user = pd.merge(user, rating[['User_Id']], how='right', on='User_Id').drop_duplicates().sort_values('User_Id')
user.head()
```

```python
# Melakukan merge antara df user dan rating untuk mendapatkan kolom Age
new = pd.merge(user[['User_Id','Age','Location']], rating, how='right', on='User_Id').drop_duplicates().sort_values('User_Id')
new.head(20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

new_place = place
CountV = CountVectorizer()

CountV.fit(new_place['Category'])
```

```python
CountV.vocabulary_.keys()
```

```python
CountV_matrix = CountV.transform(new_place['Category'])
```

```python
pd.DataFrame(
    CountV_matrix.todense(),
    columns=list(CountV.vocabulary_.keys()),
    index = new_place.Place_Name
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(CountV_matrix)
```

```python
cosine_sim_df = pd.DataFrame(cosine_sim,index=new_place['Place_Name'],columns=new_place['Place_Name'])

def simmilar_place_recommendations(place_name, csim=cosine_sim_df, param=new_place[['Place_Id','Place_Name', 'Category']], k=5):
    index = csim.loc[:, place_name].values.argpartition(range(-1, -k, -1))
    res = csim.columns[index[-1:-(k+2):-1]].tolist()
    if place_name in res:
        res.remove(place_name)
    result = param['Place_Name'].isin(res)
    result = param[result].head(2)
    print(result['Place_Id'].values)
    return result['Place_Id'].values
```

```python
simmilar_place_recommendations("Gua Pawon")
```
